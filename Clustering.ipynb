## Clustering Theoretical Questions

1. What is unsupervised learning in the context of machine learning ?

    Unsupervised learning is a type of machine learning where the model is trained on unlabeled data, meaning the data does not have predefined output labels or target values.

    In simpler terms, the algorithm tries to find hidden patterns, structures, or relationships within the data without being told what the correct output is.

2. How does K-Means clustering algorithm work ?

    **K-Means clustering** is an **unsupervised learning algorithm** used to group similar data points into **K distinct clusters** based on their features. It works by minimizing the distance between data points and their assigned cluster centers (called **centroids**).


    ### Step-by-Step Process

    1. **Choose the number of clusters (K)**
       You decide how many clusters you want the data to be divided into (e.g., K = 3).

    2. **Initialize the centroids**
       Randomly select K points from the dataset as the initial cluster centers (centroids).

    3. **Assign points to the nearest centroid**
       For each data point, calculate its distance (usually Euclidean distance) to each centroid and assign it to the cluster with the closest centroid.

    4. **Update the centroids**
       Once all points are assigned, compute the new centroid of each cluster by taking the **mean** of all data points in that cluster.

    5. **Repeat steps 3 and 4**
       Reassign points and update centroids until the centroids **stop moving significantly** or the cluster assignments no longer change.
       (This means the algorithm has **converged**.)

    6. **Output the clusters**
       The final result is K clusters with their respective centroids.

3. Explain the concept of a dendrogram in hierarchical clustering ?

    A **dendrogram** is a **tree-like diagram** used in **hierarchical clustering** to visually represent how data points are grouped together at different levels of similarity or distance.

    It shows the process of **merging (in agglomerative clustering)** or **splitting (in divisive clustering)** clusters step by step, helping us decide the **optimal number of clusters**.

4. What is the main difference between K-Means and Hierarchical Clustering ?

    The **main difference between K-Means and Hierarchical Clustering** lies in **how they form clusters** and **how the number of clusters is determined**.


    | Feature            | K-Means Clustering      | Hierarchical Clustering      |
    | ------------------ | ----------------------- | ---------------------------- |
    | Type               | Partition-based         | Hierarchical (tree-based)    |
    | Number of Clusters | Must be specified (K)   | Decided from dendrogram      |
    | Algorithm Nature   | Flat clustering         | Builds hierarchy of clusters |
    | Output             | Cluster assignments     | Dendrogram                   |
    | Cluster Shape      | Spherical               | Arbitrary                    |
    | Scalability        | Suitable for large data | Suitable for small data      |
    | Initialization     | Random centroids        | Not required                 |


5. What are the advantages of DBSCAN over K-Means ?


    **1. No Need to Specify the Number of Clusters:**
    Unlike K-Means, which requires you to predefine the number of clusters (*K*), DBSCAN automatically determines the number of clusters based on data density. This makes it more flexible, especially when you don’t know how many clusters exist in your data.

    **2. Ability to Find Arbitrary-Shaped Clusters:**
    K-Means works best for spherical clusters, but DBSCAN can discover clusters of any shape — circular, elongated, or irregular — since it relies on density rather than distance from a centroid.

    **3. Robust to Noise and Outliers:**
    K-Means assigns every point to a cluster, making it sensitive to outliers. DBSCAN, however, identifies and separates low-density points as **noise**, preventing them from distorting the cluster structure.

    **4. Handles Clusters of Different Sizes and Densities:**
    K-Means assumes clusters have similar sizes and densities. DBSCAN can handle clusters that vary in density and size, making it more suitable for real-world, unevenly distributed data.

    **5. Deterministic Results:**
    K-Means starts with random centroids, which can lead to different results on different runs. DBSCAN’s output is deterministic — it gives the same clusters every time with the same parameters.


6. When would you use Silhouette Score in clustering ?

    The **Silhouette Score** is used in clustering to **evaluate how well data points are grouped into clusters**. It measures both:

    1. **Cohesion** – how close a data point is to other points in its own cluster.
    2. **Separation** – how far a data point is from points in other clusters.

    The score ranges from **-1 to +1**:

    * **+1** → Data point is well-matched to its own cluster and far from others (good clustering).
    * **0** → Data point is on or near the boundary between clusters.
    * **-1** → Data point may have been assigned to the wrong cluster.


    ###  **When to Use Silhouette Score**

    1. **To Evaluate Clustering Quality:**
       After performing clustering (like K-Means, DBSCAN, or hierarchical), the Silhouette Score helps determine whether the clusters are **tight and well-separated**.

    2. **To Determine the Optimal Number of Clusters (K):**
       In algorithms like K-Means, you can calculate the Silhouette Score for different values of K. The value of K with the **highest average Silhouette Score** is usually considered the best choice.

    3. **To Compare Different Clustering Algorithms:**
       You can use it to compare the performance of multiple clustering methods on the same dataset. A higher Silhouette Score indicates better-defined clusters.

7. What are the limitations of Hierarchical Clustering ?

    **Hierarchical clustering** is a useful clustering method, but it has several limitations that you should be aware of:


    ### **1. High Computational Complexity**

    * The algorithm has **O(n²) or O(n³) time complexity** depending on implementation.
    * This makes it **slow and memory-intensive** for large datasets.


    ### **2. Sensitive to Noise and Outliers**

    * Outliers can create **small clusters** or distort the dendrogram.
    * Hierarchical clustering does not have a built-in mechanism to handle noise effectively.


    ### **3. Irreversibility of Merges/Splits**

    * In **agglomerative clustering**, once two clusters are merged, the decision **cannot be undone**.
    * Similarly, in **divisive clustering**, once a split occurs, it cannot be reversed.
    * This may lead to **suboptimal clustering** if early merges or splits are incorrect.


    ### **4. Choice of Distance Metric and Linkage Method**

    * Results are highly sensitive to:

      * **Distance metric** (Euclidean, Manhattan, etc.)
      * **Linkage method** (single, complete, average)
    * Different choices can lead to **very different dendrograms and clusters**.


    ### **5. Difficulty with Large Datasets**

    * Because it requires computing a **distance matrix** of all points, it becomes impractical for **datasets with thousands or millions of points**.


    ### **6. No Clear Objective Function**

    * Unlike K-Means, hierarchical clustering does not minimize a specific cost function.
    * This makes it **harder to evaluate** and optimize clustering results quantitatively.

8. Why is feature scaling important in clustering algorithms like K-Means ?

    **Feature scaling** is crucial in clustering algorithms like **K-Means** because the algorithm relies on **distance calculations** (usually Euclidean distance) to assign points to clusters. If the features have different scales, it can distort these distances and lead to misleading clusters.

    Here’s why:


    ### 1. **Equal Contribution of Features**

    * K-Means treats each feature equally when computing distances.
    * If one feature has a much larger range (e.g., 0–10000) than another (e.g., 0–10), the large-scale feature **dominates the distance calculation**, making smaller-scale features almost irrelevant.

    ### 2. **Prevents Bias in Cluster Formation**

    * Without scaling, clusters may form primarily based on the features with **larger numeric ranges**, even if other features are more meaningful.
    * Scaling ensures that **all features contribute fairly** to cluster assignment.

    ### 3. **Improves Convergence**

    * Feature scaling (like **standardization** or **normalization**) often helps K-Means **converge faster** and find a better solution because the algorithm’s optimization process is more balanced across all features.

    ### 4. **Facilitates Interpretation**

    * After scaling, clusters reflect patterns across **all features**, not just the ones with large values, making the results more interpretable and meaningful.


    ### **Common Scaling Techniques**

    * **Min-Max Normalization:** Scales features to a fixed range, e.g., 0–1.
    * **Standardization (Z-score):** Centers data at mean 0 with standard deviation 1.

9.  How does DBSCAN identify noise points ?

    **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)** identifies **noise points** based on **density criteria**. The key idea is that **dense regions form clusters**, while points in sparse regions are considered **noise**.

    Here’s how it works step by step:


    ### 1. **Core Concepts in DBSCAN**

    * **Epsilon (ε):** The radius around a point to look for neighbors.
    * **MinPts:** Minimum number of points required within ε-radius to form a dense region (core point).

    Points are classified as:

    1. **Core Point:** Has at least `MinPts` points within ε.
    2. **Border Point:** Has fewer than `MinPts` points within ε, but is within ε of a core point.
    3. **Noise Point (Outlier):** Neither a core point nor a border point.


    ### 2. **Identifying Noise Points**

    * DBSCAN examines each point in the dataset:

      1. If the point is **not a core point** (fewer than `MinPts` neighbors within ε)
      2. And it is **not reachable** from any core point (not within ε of a core point)
    * Then the point is classified as **noise**.

    In other words, **noise points lie in low-density regions** where no cluster is formed.


    ### 3. **Example**

    Suppose `MinPts = 5` and `ε = 2` units:

    * A point with 6 neighbors within distance 2 → **core point** (part of a cluster)
    * A point with 3 neighbors but near a core point → **border point** (part of cluster)
    * A point with 1 neighbor and far from any core point → **noise point** (outlier)

10. Define inertia in the context of K-Means ?


    **Inertia** is defined as the **sum of squared distances between each data point and the centroid of the cluster it belongs to**.


    [
    \text{Inertia} = \sum_{i=1}^{K} \sum_{x \in C_i} | x - \mu_i |^2
    ]

    Where:

    * (K) = number of clusters
    * (C_i) = cluster (i)
    * (x) = data point in cluster (i)
    * (\mu_i) = centroid of cluster (i)

    ### **Key Points**

    1. **Lower inertia** → points are **closer to their centroids**, meaning tighter clusters.
    2. **Higher inertia** → points are **spread out**, meaning looser clusters.
    3. Inertia is often used in the **Elbow Method** to determine the **optimal number of clusters (K)**.

11. What is the elbow method in K-Means clustering ?

    The **Elbow Method** is a technique used in **K-Means clustering** to determine the **optimal number of clusters (K)** in a dataset. It helps identify the point at which adding more clusters does **not significantly improve the clustering quality**.


    ### **How It Works**

    1. **Run K-Means for different values of K** (e.g., K = 1 to 10).
    2. **Calculate the inertia** (sum of squared distances of points to their cluster centroids) for each K.
    3. **Plot K vs. inertia** on a graph.
    4. **Look for the "elbow" point** on the curve:

       * The elbow is where the inertia **starts decreasing more slowly** as K increases.
       * This point represents the **optimal number of clusters**, balancing cluster compactness and simplicity.

12. Describe the concept of "density" in DBSCAN ?

    In **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**, the concept of **density** is central to how the algorithm identifies clusters and noise. Density refers to the **number of data points within a specific neighborhood around a point**.


13.  Can hierarchical clustering be used on categorical data ?


  **hierarchical clustering can be used on categorical data**, but it requires some adjustments because standard hierarchical clustering typically relies on **distance metrics like Euclidean distance**, which are not suitable for categorical variables.

14. What does a negative Silhouette Score indicate ?

    A **negative Silhouette Score** indicates that a data point is **likely assigned to the wrong cluster**.

    A negative Silhouette Score signals that a data point is **more similar to a different cluster than the one it was assigned**, pointing to potential issues in clustering or the need for a different number of clusters.


15. Explain the term "linkage criteria" in hierarchical clustering.

    In **hierarchical clustering**, the term **“linkage criteria”** refers to the **rule used to measure the distance between clusters** when deciding which clusters to merge (in agglomerative clustering) or split (in divisive clustering). It determines **how the proximity of clusters is calculated** based on the distances between their points.


    ### **Common Linkage Criteria**

    1. **Single Linkage (Nearest Neighbor):**

       * Distance between two clusters = **minimum distance between any pair of points** (one from each cluster).
       * Tends to produce **long, “chain-like” clusters**.

    2. **Complete Linkage (Furthest Neighbor):**

       * Distance between two clusters = **maximum distance between any pair of points**.
       * Produces **more compact, evenly shaped clusters**.

    3. **Average Linkage:**

       * Distance between two clusters = **average of all pairwise distances** between points in the clusters.
       * Balances the tendencies of single and complete linkage.

    4. **Centroid Linkage:**

       * Distance between two clusters = **distance between their centroids** (mean of points in each cluster).

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities ?

    **K-Means clustering can perform poorly on data with varying cluster sizes or densities** because of the way it assigns points to clusters based on **distance to centroids**. Here’s why:


    ### **1. Assumes Equal Cluster Sizes**

    * K-Means assumes that clusters are **roughly equal in size**.
    * If one cluster is much larger or smaller than another, the centroid may **not accurately represent all points**, causing misclassification.


    ### **2. Sensitive to Cluster Density**

    * K-Means assigns points to the **nearest centroid** without considering density.
    * Dense clusters with many points can “pull” centroids toward them, while sparse clusters may get **underrepresented**.
    * Points in low-density clusters may be incorrectly assigned to high-density clusters nearby.


    ### **3. Poor Handling of Non-Spherical Clusters**

    * K-Means works best for **spherical, evenly sized clusters**.
    * If clusters are elongated, irregularly shaped, or have different densities, the distance-to-centroid criterion fails to separate them correctly.


    ### **4. Outlier Sensitivity**

    * Sparse clusters are more susceptible to being **dominated by nearby outliers or denser clusters**, further degrading performance.



17. What are the core parameters in DBSCAN, and how do they influence clustering ?

    In **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)**, the clustering results are mainly determined by **two core parameters**:


    ### **1. Epsilon (ε) – Neighborhood Radius**

    * Defines the **radius around a point** to search for neighboring points.
    * Determines what is considered “close” or part of the same dense region.

    **Effect on Clustering:**

    * **Small ε:** May result in many points being labeled as **noise**, producing smaller, tighter clusters.
    * **Large ε:** May merge distinct clusters into one, producing **fewer, larger clusters**.


    ### **2. MinPts – Minimum Points**

    * The **minimum number of points required** within the ε-radius to form a **core point** (dense region).

    **Effect on Clustering:**

    * **Small MinPts:** More points qualify as core points, possibly forming **more clusters** and including sparse regions.
    * **Large MinPts:** Fewer points qualify as core points, leading to **larger clusters** and more points classified as **noise**.


    ### **How These Parameters Work Together**

    * **Core points:** Points with ≥ MinPts neighbors within ε.
    * **Border points:** Points within ε of a core point but with < MinPts neighbors.
    * **Noise points:** Points that are neither core nor border points.

    Choosing ε and MinPts carefully is **critical** to capturing the true cluster structure in the data. Often, a **k-distance graph** is used to help determine an appropriate ε.

18. How does K-Means++ improve upon standard K-Means initialization ?

    **K-Means++** is an improvement over standard K-Means that focuses on **better initialization of centroids** to improve clustering performance and convergence.


    ### **Problem with Standard K-Means Initialization**

    * Standard K-Means selects **initial centroids randomly**.
    * Poorly chosen initial centroids can lead to:

      * Slow convergence
      * Suboptimal clusters (higher inertia)
      * Different results in different runs due to randomness


19. What is agglomerative clustering ?

    **Agglomerative clustering** is a type of **hierarchical clustering** that follows a **bottom-up approach** to build clusters.

    Agglomerative clustering is a **bottom-up hierarchical method** where each point starts as its own cluster, and clusters are **iteratively merged based on similarity** to form a hierarchy of clusters.


20. What makes Silhouette Score a better metric than just inertia for model evaluation?

    The **Silhouette Score** is often considered a better metric than **inertia** for evaluating clustering because it provides a **more comprehensive assessment of cluster quality**. Here’s why:


    ### **1. Considers Both Cohesion and Separation**

    * **Inertia** measures only **cohesion** — how close points are to their cluster centroids.
    * **Silhouette Score** considers both:

      * **Cohesion (a):** How close a point is to other points in its own cluster.
      * **Separation (b):** How far a point is from points in the nearest neighboring cluster.
    * This means Silhouette Score evaluates not just tightness, but also **how well clusters are separated**, giving a fuller picture of clustering quality.


    ### **2. Scales Across Different Numbers of Clusters**

    * Inertia **always decreases as K increases**, making it hard to compare models with different numbers of clusters.
    * Silhouette Score ranges between **-1 and +1**, allowing for **comparison across different K values**.

      * +1 → well-clustered
      * 0 → overlapping clusters
      * -1 → misclassified points

    ### **3. Detects Misclassified Points**

    * Silhouette Score provides a **per-point measure**, highlighting points that may be **assigned to the wrong cluster**.
    * Inertia does not give information at the individual point level; it only sums squared distances globally.

    ### **4. Helps Decide Optimal K More Reliably**

    * Using inertia alone (Elbow Method) can be ambiguous if the “elbow” is not clear.
    * Silhouette Score directly indicates **the K that maximizes overall cluster quality**, considering both cohesion and separation.

```python
#21.  Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y_true = make_blobs(n_samples=400, centers=4, cluster_std=1.0, random_state=42)

kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

plt.figure(figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50, alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, marker='X', label='Centroids')
plt.title('K-Means Clustering with 4 Centers')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
#23. Generate synthetic data using make_moons and apply DBSCAN. Highlight outliers in the plot.


import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import numpy as np


X, y_true = make_moons(n_samples=300, noise=0.05, random_state=42)

dbscan = DBSCAN(eps=0.2, min_samples=5)
y_pred = dbscan.fit_predict(X)

core_samples_mask = np.zeros_like(y_pred, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True
outliers_mask = (y_pred == -1)

plt.figure(figsize=(8,6))

plt.scatter(X[~outliers_mask, 0], X[~outliers_mask, 1], c=y_pred[~outliers_mask],
            cmap='viridis', s=50, alpha=0.7, label='Clustered Points')

plt.scatter(X[outliers_mask, 0], X[outliers_mask, 1], c='red', s=80, marker='x', label='Outliers')

plt.title('DBSCAN Clustering on make_moons Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
#24. Load the Wine dataset and apply K-Means clustering after standardizing the features. Print the size of each cluster.

from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

wine = load_wine()
X = wine.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X_scaled)

(unique, counts) = np.unique(y_kmeans, return_counts=True)
cluster_sizes = dict(zip(unique, counts))
print("Cluster sizes:", cluster_sizes)
```

```python
#28.  Load the Digits dataset, reduce it to 2D using PCA, and visualize clusters from K-Means.

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

digits = load_digits()
X = digits.data
y_true = digits.target

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X)

kmeans = KMeans(n_clusters=10, random_state=42)
y_kmeans = kmeans.fit_predict(X_pca)

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans, cmap='tab10', s=50, alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, marker='X', label='Centroids')
plt.title('K-Means Clusters on 2D PCA-reduced Digits Data')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.show()
```

```python
#29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X, y_true = make_blobs(n_samples=400, centers=4, cluster_std=1.0, random_state=42)

k_values = range(2, 6)
sil_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    y_kmeans = kmeans.fit_predict(X)
    score = silhouette_score(X, y_kmeans)
    sil_scores.append(score)
    print(f"Silhouette Score for k={k}: {score:.4f}")

plt.figure(figsize=(8,5))
plt.bar(k_values, sil_scores, color='skyblue')
plt.xticks(k_values)
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for Different k Values')
plt.show()
```

```python
#31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y_true = make_blobs(n_samples=400, centers=3, cluster_std=2.0, random_state=42)  # higher std for overlap

kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)

h = 0.1  # step size for mesh
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='viridis')
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', s=200, marker='X', label='Centroids')
plt.title('K-Means Clustering with Overlapping Blobs and Decision Boundaries')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
#34.Load the Breast Cancer dataset and compare inertia values for K = 2 to 6 using K-Means. Show results in a line plot.

import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

cancer = load_breast_cancer()
X = cancer.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

k_values = range(2, 7)
inertia_values = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia_values.append(kmeans.inertia_)
    print(f"Inertia for k={k}: {kmeans.inertia_:.2f}")

plt.figure(figsize=(8,5))
plt.plot(k_values, inertia_values, marker='o', linestyle='-', color='blue')
plt.xticks(k_values)
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('K-Means Inertia for Different K (Breast Cancer Dataset)')
plt.show()
```

```python
#44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.

import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

iris = load_iris()
X = iris.data

kmeans = KMeans(n_clusters=3, random_state=42)
y_kmeans = kmeans.fit_predict(X)

sil_values = silhouette_samples(X, y_kmeans)
avg_score = silhouette_score(X, y_kmeans)
print(f"Average Silhouette Score: {avg_score:.4f}")

y_sorted = np.argsort(y_kmeans)
sil_sorted = sil_values[y_sorted]
cluster_sorted = y_kmeans[y_sorted]

plt.figure(figsize=(10,6))
plt.bar(range(len(sil_sorted)), sil_sorted, color=plt.cm.viridis(cluster_sorted / 3))
plt.axhline(avg_score, color='red', linestyle='--', label=f'Average Silhouette Score ({avg_score:.2f})')
plt.xlabel('Sample index (sorted by cluster)')
plt.ylabel('Silhouette Coefficient')
plt.title('Silhouette Coefficient for Each Sample (Iris Dataset, K-Means)')
plt.legend()
plt.show()
```
